```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('../../files/trees.csv')
base.head()
```

```python
plt.boxplot(base.Volume, vert=False, showfliers=True, notch=False, patch_artist=True)
plt.title("Trees")
plt.xlabel("Volume")
```

```python
plt.boxplot(base)
plt.title("Trees")
plt.xlabel("data")
```

```python
plt.boxplot(base.Volume, vert=True)
plt.boxplot(base.Girth, vert=True)
plt.boxplot(base.Height, vert=True)
plt.title("Trees")
plt.xlabel("Data")
```
