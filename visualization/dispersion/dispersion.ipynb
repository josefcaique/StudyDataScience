```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
plt.hist(base.iloc[:1], bins=6)
```

```python
sns.histplot(base.iloc[:,1], kde=False, bins=6, color='blue').set(title="Trees")
```

```python
sns.kdeplot(base.iloc[:,1], color='blue').set(title='Trees')
```

```python
sns.histplot(base.iloc[:,1], kde=True, bins=6, color='blue').set(title='Trees')
```
